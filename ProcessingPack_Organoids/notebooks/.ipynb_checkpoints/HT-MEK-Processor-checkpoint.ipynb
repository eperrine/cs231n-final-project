## Description
<br>

**Stitched-image processing for MITOMI-type experiment**
 1. Stitched Image "stamping"
 2. Chamber and Button Finding (using Hough transform and grid-search intensity maximization, respectively)
 3. Chamber and Button Intensity Quantification
 4. Report generation (Pandas DataFrames and CSVs)
 
**Classes for processing single chip images and multiple chip images**
 1. ChipSeries (parent)
     1. StandardSeries (ChipSeries): eMITOMI-Specific Chamber Standard Curve
     2. Timecourse (ChipSeries): Generic Timecourse Image Processing
     3. Titration (ChipSeries)
 2. ChipQuant (parent)
 3. Assay (parent)
     1. TurnoverAssay (Assay)
 4. AssaySeries(parent)
