### 1. Read and rotate image, find boundaries for partitioning into sub-arrays

```python
# oc = ((2555, 4814), (10158, 3660), (4212, 16545), (11869, 15417)) # Day 6 well 1 (0), subarray_dims = (2, 3)
oc = ((4887, 430), (16658, 1055), (3949, 16266), (15755, 16957)) # Day 6 well 2 (1), subarray_dims = (3, 4)

subarray_dims = (3, 4)
srcHandle = Path('/media/Scratch/Research/Github_SupportFiles/TestImaging/Stacked_Imgs_Huge/20181210_multWT_day6/day6_Tiles/well_1.tif')
targetHandle = Path('/media/Scratch/Research/Github_SupportFiles/TestImaging/Stacked_Imgs_Huge/20181210_multWT_day6/day6_Tiles/well_1_rotated.tif')

rot_img = readAndRotateImg(srcHandle, oc, subarray_dims, targetHandle = targetHandle)
rot_corners = transformCorners(readTiff(srcHandle), rot_img, oc)
divisions = getPartitions(rot_corners, subarray_dims)
```

### 3. Perform Processing

```python
from datetime import datetime

tile_dims = (20, 20)
channel = '2bf'
exposure = 10

date = datetime(2019, 1, 13)
well = 2

targetHandle = Path('/media/Scratch/Research/Github_SupportFiles/TestImaging/Stacked_Imgs_Huge/20181210_multWT_day6/day6_Tiles/well_1_rotated.tif')
arrayRepo = processTiles(str(targetHandle), e, pinlist, divisions, tile_dims, subarray_dims, channel, exposure, date = date, well = well, desc = e.info)

# stamped_img_path = Path('/media/Scratch/Research/Github_SupportFiles/TestImaging/20181023_WT_P_APE_early_day0/4230WT_P_APE/stamped_4230_WT_day0_stitched.tif')
# skimage.external.tifffile.imsave(stamped_img_path, skimage.img_as_uint(chip.summary_image('none')))
```

### 4. Summarize the data as a CSV and Stacked Image

```python
outRoot = Path('/media/Scratch/Research/Github_SupportFiles/TestImaging/Stacked_Imgs_Huge/20181210_multWT_day6/day6_Tiles/') #Folder to write summary image and CSV

ar = arrayRepo[0, 0] #Need any sub-array to generate the hash, so the first will do
hashStr = generateWellIdentifier(ar)
summary_imgPath = outRoot.joinpath('{}.tif'.format(hashStr))
summary_csvPath = outRoot.joinpath('{}.csv'.format(hashStr))

summaryDF = summarizeArrayRepo(arrayRepo)
summaryDF.to_csv(summary_csvPath)
writeSubArraySummaryImg(arrayRepo, summary_imgPath)
```

```python
# Specify the indices

subarray_xindex = 3
subarray_yindex = 2
xindex = 1
yindex = 3

# Gather the well data
sdf = summaryDF.set_index(['subarray_xindex', 'subarray_yindex', 'x', 'y']).loc[(subarray_xindex, subarray_yindex)]
blank_summary = arrayRepo[subarray_xindex, subarray_yindex].summary_image('blank')
td = sdf.loc[(xindex, yindex)]
xslice = slice(*td.summaryImg_xslice)
yslice = slice(*td.summaryImg_yslice)
slicedImg = blank_summary[xslice, yslice]

# Show the well
pl.imshow(slicedImg)
```
