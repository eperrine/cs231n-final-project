### 4. Specify the slices for each well, save each well image separately

```python
well_indices = range(2)
sf = 5 #scale factor
wellslices = ((slice(0, 4500*sf),slice(0, 3100*sf)), None)

indices_slices = dict(zip(well_indices, wellslices))
common_name = Path('/media/Scratch/Research/Github_SupportFiles/TestImaging/Stacked_Imgs_Huge/20181212_multWT_day8/')

for index, slices in indices_slices.items():
    savePath = common_name.joinpath(Path('well_{}.tif'.format(str(index))))
    external.tifffile.imsave(str(savePath), img[slices[0], slices[1]], compress = 9)
    break
```
