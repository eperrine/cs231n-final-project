## Imports

```python
import sys
from pathlib import Path
import matplotlib.pyplot as pl
sys.path.insert(0, '/home/dmokhtari/Documents/GithubRepos/ImageStitcher')
from imagestitcher import orgstitcher, stitcher

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
%load_ext autoreload
%autoreload 2
```

### 1. Establish a DataFrame with the associated paths, get possible image dimensions
>Without metadata, we don't know how many rows/columns were in the original raster. So, it's necessary to guess dimensions
using the total number of images and trying a reasonable pair of dimensions for stitching.

```python
#Path containing each of the .tif rastered images
path = Path('/media/Scratch/Research/Github_SupportFiles/TestImaging/Stacked_Imgs_Huge/20181210_multWT_day6/multWT')

rasterData = orgstitcher.makePathRecords(path)
display(rasterData.head())
print('Possible Dimensions: \n{}'.format(orgstitcher.getFactors(rasterData)))
```

### 2. Set up the raster, using a test number images per row from the list of possible factors
> Using our "guessed" number of columns, set of the other stitching parameters including the %overlap (30% by default), and provide other dummy parameters needed to initialize
we also need to provide the dimensions of the individual images

```python
ncols = 57 #tilesPerHorizontalStrip

params = stitcher.RasterParams(path, 'ipnb', 0.3, [100], '2bf', 1, autoFF = False) #Only overlap should be changed. Some parameters are unused 
params.dims = (1, ncols) #This is the row dimension
params.imsize = (1920, 1440) #Pixel dimension of the individual image
```

### 3. Stitch the image
>Perform the image stitching. This will take ~16 minutes.

```python
img = orgstitcher.stitchOrgRaster(rasterData, params, ncols)
```

### 4. Save the stitched img
> Specify the path of the file to be written

```python
savePath = '/media/Scratch/Research/Github_SupportFiles/TestImaging/Stacked_Imgs_Huge/20181210_multWT_day6/fullStitch_compressed.tif'
orgstitcher.saveRaster(img, savePath)
```

### 5. Read back in the image and resize it 1:5 so that corner positions can be specified
> Before reading  back in the image, it's recommended to reset the kernel

```python
path = '/media/Scratch/Research/Github_SupportFiles/TestImaging/Stacked_Imgs_Huge/20181210_multWT_day6/fullStitch_compressed.tif'
img = orgstitcher.readImage(path)
```

```python
resizeFactor = 5
smallImg = orgstitcher.resizeImage(img, resizeFactor)
```

### 6a. Visualize the resized image
> Display the rescaled image and use this image to  make note of _rough_ corner positions for each of the subarrays. 
Note those corner positions in *7.* below, in order from top-left->top-right, row by row (i.e., well 0 is  top  left, well 3 is middle left, well 6 is bottom left)

```python
pl.figure(figsize = (8, 8))
pl.imshow(smallImg)
pl.xticks(range(0, smallImg.shape[1], 1000), rotation='vertical');
pl.yticks(range(0, smallImg.shape[0], 1000));
```

### 6b: Save resized img if needed
> Not required, but if you want to  view the full image with ImageJ you'll want to do this.

```python
savePath = '/home/dmokhtari/Downloads/compressed_testimg.tif'
orgstitcher.saveRaster(smallImg, savePath)
```

### 7. Specify the slices for each well, save each well image separately
> Specify the corner positions of each of the wells (as described in 6a.), and specify the folder in which to export the well sub-images

```python
maxwidth = smallImg.shape[1]
maxheight = smallImg.shape[0]

wellslices = (((0, 4500), (0, 3100)), 
              ((0, 4500), (5500, 10000)), 
              ((0, 4500), (12000, maxwidth)),
              ((6500, 11500), (0, 3100)), 
              ((6500, 11500), (5500, 10000)),
              ((6500, 11500), (12000, maxwidth)),
              ((13500, maxheight), (0, 3100)),
              ((13500, maxheight), (5500, 10000)),
              ((13500, maxheight), (12000, maxwidth)))

exportPath = Path('/media/Scratch/Research/Github_SupportFiles/TestImaging/Stacked_Imgs_Huge/20181210_multWT_day6/day6_Tiles')
orgstitcher.sliceImage(wellslices, exportPath, smallImg, img, 5, 'fullimg slicing test')
```
