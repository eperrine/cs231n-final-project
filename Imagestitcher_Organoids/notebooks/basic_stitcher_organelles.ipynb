### 1. Establish a DataFrame with the associated paths, get possible image dimensions
>Without metadata, we don't know how many rows/columns were in the original raster. So, it's necessary to guess dimensions
using the total number of images and trying a reasonable pair of dimensions for stitching.

```python
#Path containing each of the .tif rastered images
#path = Path('/media/Scratch/Research/Github_SupportFiles/TestImaging/Stacked_Imgs_Huge/20181210_multWT_day6/multWT')
path = Path('/Volumes/ALEX_SSD/PDAC_drugtest/day19/trypan/PDAC2_B1')


rasterData = orgstitcher.makePathRecords(path)
display(rasterData.head())
print('Possible Dimensions: \n{}'.format(orgstitcher.getFactors(rasterData)))
```

### 2. Set up the raster, using a test number images per row from the list of possible factors
> Using our "guessed" number of columns, set of the other stitching parameters including the %overlap (30% by default), and provide other dummy parameters needed to initialize
we also need to provide the dimensions of the individual images

```python
ncols = 17 #tilesPerHorizontalStrip

params = stitcher.RasterParams(path, 'ipnb', 0.3, [100], '2bf', 1, autoFF = False) #Only overlap should be changed. Some parameters are unused 
params.dims = (1, ncols) #This is the row dimension
params.imsize = (1920, 1440) #Pixel dimension of the individual image
```

### 3. Stitch the image
>Perform the image stitching. This will take ~16 minutes.

```python
img = orgstitcher.stitchOrgRaster(rasterData, params, ncols)
```

### 4. Save the stitched img
> Specify the path of the file to be written

```python
savePath = '/Users/alexandrasockell/Desktop/Organoids/Data/stitched/PDAC2_B1/day19_stitched.tif'
orgstitcher.saveRaster(img, savePath)
```
